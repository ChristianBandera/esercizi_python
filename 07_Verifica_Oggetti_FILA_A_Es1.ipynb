**ESERCIZIO 1 – MACCHINE E GARAGE**

Creare la classe `Garage` per contenere una serie di oggetti `Macchina`.

La classe `Garage` deve contenere:

* un attributo che rappresenti una lista di macchine parcheggiate;
* un metodo per aggiungere una macchina al garage;
* un metodo per rimuovere una macchina (dato il numero di targa);
* un metodo per visualizzare l’elenco delle macchine presenti;
* un metodo per cercare una macchina dato il numero di targa e restituire le sue informazioni.

---



````python
class Macchina:
    def __init__(self, targa, marca, modello, anno):
       self.targa=targa
       self.marca=marca
       self.modello=modello
       self.anno=anno

    def __str__(self):
        return f"\nTarga: {self.targa} Marca: {self.marca} Modello: {self.modello} Anno: {self.anno}"


class Garage:
    def __init__(self):
        self.garage=[]

    def aggiungi_macchina(self, macchina):
        self.garage.append(macchina)
        print("Macchina aggiunta")

    def rimuovi_macchina(self, targa):
        print(f"Rimozione della macchina con targa {targa}")
        trovato=False
        indice=-1
        for macchina in self.garage:
            if(macchina.targa==targa):
                trovato=True
                indice= self.garage.index(macchina)
                self.garage.pop(indice)
                print(f"{macchina} rimossa")
        if(trovato==False):
            print("Macchina non trovata")

    def elenco_macchine(self):
        print("Garage: ")
        for macchina in self.garage:
            print(macchina)

    def cerca_macchina(self, targa):
        print(f"Ricerca della macchina con targa {targa}")
        trovato=False
        for macchina in self.garage:
            if(macchina.targa==targa):
                trovato=True
                return macchina
        if(trovato==False):
            print("Macchina non trovata")
        return -1
    

#output
garage=Garage()

while(True):
    print("0.Esci \n1.Aggiungi macchina \n2.Rimuovi macchina \n3.Visualizza garage \n4.Cerca Macchina")
    print()
    scelta=int(input("Inserisci scelta: "))
    print()
    if scelta==0:
        break
    elif scelta==1:
        targa=input("Inserisci targa: ")
        while(targa=="" or len(targa)!=7):
            targa=input("Inserisci targa: ")
        marca=input("Inserisci marca: ")
        while(marca==""):
            marca=input("Inserisci marca: ")
        modello=input("Inserisci modello: ")
        while(modello==""):
            modelo=input("Inserisci modello: ")
        anno=int(input("Inserisci anno: "))
        while(anno=="" or anno<0):
            anno=input("Inserisci anno: ")
        macchina=Macchina(targa, marca, modello, anno)
        garage.aggiungi_macchina(macchina)
        print()
    elif scelta==2:
        targa=input("Inserisci targa: ")
        while(targa=="" or len(targa)!=7):
            targa=input("Inserisci targa: ")
        garage.rimuovi_macchina(targa)
        print()
    elif scelta==3:
        garage.elenco_macchine()
    elif scelta==4:
        targa=input("Inserisci targa: ")
        while(targa=="" or len(targa)!=7):
            targa=input("Inserisci targa: ")
        print(garage.cerca_macchina(targa))
        print()
    else:
        print("Scelta errata")
        print()
````
