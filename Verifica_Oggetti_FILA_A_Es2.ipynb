**ESERCIZIO 2 : PAGELLE**

Crea una classe che rappresenti una **pagella scolastica**. La classe accetta in input una lista di numeri (voti) che rappresentano le valutazioni in ciascuna materia.

La classe deve avere queste funzionalità:

1. Stampando un oggetto `Pagella` il risultato sarà la lista dei voti *(implementare `__repr__`)*

2. La classe deve avere il metodo `media()`: ritorna la media aritmetica dei voti della pagella


3. La classe deve supportare l’**indexing**, ossia indicando l’indice si ottiene il voto corrispondente *(implementare `__getitem__`)*

   ```python
   p = Pagella([6, 8, 7])
   print(p[1])  # stampa 8
   ```

4. Confrontando due oggetti `Pagella`, il risultato sarà `True` se contengono esattamente gli stessi voti nello stesso ordine, `False` altrimenti *(implementare `__eq__`)*

5. Sottraendo una `Pagella` da un'altra si ottiene una nuova `Pagella` con la differenza tra i voti nelle stesse posizioni *(implementare `__sub__`)*


6. La classe deve avere il metodo `impegno()`: ritorna una misura dell’impegno complessivo calcolata come **radice quadrata del prodotto scolastico** della pagella con sé stessa
   *(es. `Pagella([6, 8])` → `√(6² + 8²) = √(100) = 10.0`)*

**Nota**: Per i punti 4, 5 le pagelle devono avere la **stessa dimensione**. In caso contrario, stampare il messaggio
> `"Le pagelle hanno un numero diverso di materie e non possono essere confrontate"`
> e restituire `None`.

---


````python
import math

class Pagella:
    def __init__(self, listaVoti):
        self.listaVoti=listaVoti

    def __repr__(self):
        return str(self.listaVoti)
    
    def media(self):
        somma=0
        num=len(self.listaVoti)
        for i in range(len(self.listaVoti)):
            somma+=self.listaVoti[i]
        if(num>0):
            return round(somma/num,2)
        else:
            print("Media non calcolabile")
            return -1

    def __getitem__(self, indice):
        for i in range(len(self.listaVoti)):
            if(i==indice):
                return self.listaVoti[i]
        return -1
    
    def __eq__(self, pagella):
        uguale=True 
        lunghezza=len(pagella.listaVoti)
        print(lunghezza)
        for i in range(len(self.listaVoti)):
            if(self.listaVoti[i]==pagella[i]):
                uguale=False
        if(len(self.listaVoti)==lunghezza):
            return uguale
        else:
            print("Le pagelle hanno un numero diverso di materie e non possono essere confrontate")
            return None
    
    def __sub__(self, pagella):
        nuovaPagella=[]
        lunghezza=len(pagella.listaVoti)
        if(len(self.listaVoti)==lunghezza):
            for a,b in zip(self.listaVoti, pagella):
                nuovaPagella.append(a-b)
            return nuovaPagella
        else:
            print("Le pagelle hanno un numero diverso di materie e non possono essere confrontate")
            return None
    
    def impegno(self):
        somma=0
        for i in range(len(self.listaVoti)):
            somma+=(self.listaVoti[i] * self.listaVoti[i])
        impegno= math.sqrt(somma)
        return impegno
    

#output
pagella1=Pagella([5,6,7,8])
pagella2=Pagella([4,5,6,7])
pagella3=Pagella([5,6])
pagella4=Pagella([5,6,7,8])

#repr
print("Visualizzazione pagelle: ")
print(f"Pagella 1: {pagella1}")
print(f"Pagella 2: {pagella2}")
print(f"Pagella 3: {pagella3}")

#media
print(f"La media dei voti di pagella1 è: {pagella1.media()}")
print(f"La media dei voti di pagella2 è: {pagella2.media()}")
print(f"La media dei voti di pagella3 è: {pagella3.media()}")
print()
#getitem
print(pagella2[2])
print()
#eq
if(pagella1==pagella2):
    print(f"Le pagelle pagella1 e pagella2 sono uguali")
if(pagella1==pagella4):
    print(f"Le pagelle pagella1 e pagella4 sono uguali")
if(pagella1==pagella3):
    print(f"Le pagelle pagella1 e pagella3 sono uguali")
print()
#sub
print("Sottrazioni")
print(f"pagella1-pagella2: {pagella1-pagella2}")
print(f"pagella1-pagella3: {pagella1-pagella3}")
print()
#impegno
print(f"Impegno di pagella3: {pagella3.impegno()}")
````
